**Objectif :** Soit $X$ et $Y$ deux s√©quences. Par exemple $X = "MAMIE"$ et $Y="MAMA\quad MIA"$.  
On souhaite:

Pour cr√©er les sous-s√©quences, on part de la s√©quence initiale.  
On commence par g√©n√©rer toutes les cha√Ænes possibles en enlevant un caract√®re.  
Pour chacune de ces sous-cha√Æne cr√©√©e, on g√©n√®re les sous-cha√Ænes ayant encore un caract√®re de moins.  
Si une sous-cha√Æne n'a pas d√©j√† √©t√© cr√©√©e, on l'ajoute √† la liste finale.  
Jusqu'√† atteindre une cha√Æne vide.

## Approche par masquage
Soit L la taille de la chaine.  
On g√©n√®re tous les masques binaires possibles sur L bits et on op√®re un masquage de la cha√Æne √† partir de ces masques:  
On ne conserve un caract√®re que s'il se retrouve en face d'un '1'.
![AND_str](https://raw.githubusercontent.com/ebuonocore/NSIT_19_Programmation_dynamique/main/sources/AND_masque.png)

```python
def AND_str(chaine:str, masque:str)->str:
    """ Op√®re une fonction de masquage entre la chaine et le masque
        Pr√©condition: La taille des deux param√®tres est identique
        1 dans le masque -> On conserve le cract√®re
        0 -> On supprime ce caract√®re
    """
    sortie = ""
    for i in range(len(chaine)):
        if masque[i]=='1':
            sortie += chaine[i]
    return sortie

def sous_sequences_masque(chaine_initiale:str):
    """ G√©n√®re la liste de toutes les sous-s√©quences possibles de la cha√Æne initiale
    Soit L = len(chaine_initiale)
    On construit tous les masques binaires possibles sur L bits 
    On effectue un traitement logique entre la cha√Æne initiale et chaque masque:

    """
    liste_sous_sequences = []
    L = len(chaine_initiale)
    for masque_dec in range(2**L):
        # Supprime les caract√®re '0b' devant le nombre binaire
        masque_bin = bin(masque_dec)[2:] 
        # Calcule le nombre de '0' manquant pour √©crire sur L bits
        Nb_zeros = L - len(masque_bin) 
        masque_bin = '0'*Nb_zeros + masque_bin # Ajoute les '0' manquants
        liste_sous_sequences.append(AND_str(chaine_initiale, masque_bin))
    return liste_sous_sequences

print (AND_str("MAMA MIA", "01100010"))
print (AND_str("MAMA MIA", "10010110"))
sous_sequences_masque("MAMIE")
```
